Increasing the vibrance of a photograph is probably achieved by applying an intensity transformation such
as
f (x) = min(x+((a/128)e^(−((x−128)^2)/(2(σ^2)))), 255)
to the saturation plane, where x is the input intensity, a ∈ [0,1] and σ = 70. [10]

(b) Apply the aforementioned intensity transformation to the saturation plane.


```python
def vibrance_intensity_transformation(x, a):
    sigma = 70
    return np.minimum(x + ((a / 128) * np.exp(-((x - 128)**2) / (2 * (sigma**2)))), 255)

# Apply the intensity transformation to the saturation plane
a = 0.5  # Adjust this value for desired effect
enhanced_saturation = vibrance_intensity_transformation(saturation, a)
```

(c) Adjust a to get a visually pleasing output. Report the value of a.


(d) Recombine the three planes.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('images\en3160_assignment_01_Page_2_Image_0002.jpg')

# Convert the image to HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Split the HSV image into hue, saturation, and value planes
hue, saturation, value = cv2.split(hsv_image)

# Apply the intensity transformation to the saturation plane
a = 0.5  # Adjust this value for desired effect
enhanced_saturation = vibrance_intensity_transformation(saturation, a)

# Combine the modified saturation with the hue and value planes
enhanced_hsv = cv2.merge([hue, enhanced_saturation, value])

# Convert the result back to BGR color space
enhanced_bgr = cv2.cvtColor(enhanced_hsv, cv2.COLOR_HSV2BGR)

# Display the images
plt.figure(figsize=(12, 8))

plt.subplot(131), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image"), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(cv2.cvtColor(enhanced_bgr, cv2.COLOR_BGR2RGB))
plt.title("Vibrance-Enhanced Image (a = {})".format(a)), plt.xticks([]), plt.yticks([])

# Plot the intensity transformation
x = np.arange(256)
transformed_x = vibrance_intensity_transformation(x, a)
plt.subplot(133), plt.plot(x, transformed_x)
plt.title("Intensity Transformation"), plt.xlabel("Input Intensity"), plt.ylabel("Output Intensity")

plt.tight_layout()
plt.show()
```
